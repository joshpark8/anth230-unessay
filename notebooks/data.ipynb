```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../data/World Bank/Gender_StatsCSV.csv')
```

```python
# Cell 2: Display basic information about the dataset
print("Dataset Info:")
print(df.info())
print("\nFirst 5 rows of the dataset:")
print(df.head())
```

```python
# Group by the country column
country_dfs = {country: data for country, data in df.groupby("Country Name")}
```

```python
for country, data in country_dfs.items():
    filename = f"../data/World Bank/countries/{country.replace(' ', '_')}_stats.csv"
    data.to_csv(filename, index=False)
```
